```python
# @title Frequency vs Age
age_groups = ['Age 16-19', 'Age 20-24', 'Age 25-34', 'Age 35-44', 'Age 45-54', 'Age 55-64', 'Age 65+']
for age_group in age_groups:
    plt.figure(figsize=(6, 3))
    plt.hist(data[age_group], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {age_group}')
    plt.xlabel('Age Group')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
# @title Year vs Age
for age_group in age_groups:
    plt.figure(figsize=(10, 6))
    data.boxplot(column=age_group, by='Year')
    plt.title(f'{age_group} by Year')
    plt.xlabel('Year')
    plt.ylabel('Age Group')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
# @title 找出資料關聯性
numeric_data = data.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numeric_data.corr() #由此可知哪些列之間具相關性
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix of Age Groups')
plt.show()
```

```python
age_group_trends = data.groupby('Year')[age_groups].mean()

plt.figure(figsize=(10, 6))
for age_group in age_groups:
    plt.plot(age_group_trends.index, age_group_trends[age_group], label=age_group)

plt.title('Trends in Age Distribution Over Time')
plt.xlabel('Year')
plt.ylabel('Mean Population')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import IsolationForest

clf = IsolationForest(contamination=0.05)
outliers = clf.fit_predict(numeric_data)
outliers  #運用IsolationForest找出異常值，並以1 -1表示之
```

```python
plt.scatter(numeric_data.iloc[:, 0], numeric_data.iloc[:, 1], c=outliers, cmap='viridis')
plt.scatter(numeric_data.iloc[outliers == -1, 0], numeric_data.iloc[outliers == -1, 1], c='red', label='Outliers')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Anomaly Detection using Isolation Forest')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
data.sort_index(inplace=True)
plt.figure(figsize=(14, 7))

for age_group in age_groups:
    # 建立和擬合 SARIMAX 模型
    model = SARIMAX(data[age_group],
                    order=(1, 1, 1),
                    seasonal_order=(1, 1, 1, 12),
                    enforce_stationarity=False,
                    enforce_invertibility=False)
    model_fit = model.fit(disp=False)  # 隱藏擬合過程的輸出信息

    # 預測未來 120 個月（10年）
    forecast = model_fit.get_forecast(steps=120)
    forecast_index = pd.date_range(start=data.index[-1], periods=121, freq='M')[1:]
    forecast_data = forecast.predicted_mean

    result_date = data.index.append(forecast_index)
    result_data = pd.concat([data[age_group], forecast_data])
    plt.plot(result_date, result_data, label=f'{age_group}')

plt.title('Forecasting Next 10 Years of Unemployment Rate by Age Group')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.legend()
plt.show()
```
